```python
def seed_everything(seed):
    random.seed(seed) #파이썬 자체 모듈 random 모듈의 시드 고정
    os.environ['PYTHONHASHSEED'] = str(seed) 
    np.random.seed(seed)
    
seed_everything(37)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

ttrain = train.copy()
ttrain = ttrain.drop(columns = ['허위매물여부'], axis = 1)
# 결측 여부를 이진 변수로 변환
for col in ttrain.columns:
    ttrain[f'{col}'] = ttrain[col].isna().astype(int)
    
ttrain.head()
```

```python
ttrain.missing_count.unique()
```

```python
ttrain.missing_count.value_counts()
```

```python
ttrain
```

```python
ttrain_ = pd.concat([ttrain, train.허위매물여부], axis = 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# missing_count와 허위매물여부 조합별 개수 계산
count_df = ttrain_.groupby(['missing_count', '허위매물여부']).size().reset_index(name='count')

# 산점도 그리기
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=count_df['missing_count'], 
    y=count_df['허위매물여부'], 
    size=count_df['count'],  # 데이터 개수를 점 크기로 적용
    sizes=(20, 300),
    alpha=0.6,
    legend=False 
)

plt.xlabel("결측 개수 (missing_count)")
plt.ylabel("허위매물 여부")
plt.title("missing_count vs 허위매물여부 산점도 (크기: 데이터 개수)")

plt.show()
```

```python
ttrain_ = ttrain_[['전용면적', '해당층', '총층', '방수', '욕실수', '총주차대수','missing_count', '허위매물여부']]
```

```python
# 결측 여부와 목표변수 간 상관관계 분석
corr_matrix = ttrain_.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".3f")
plt.title("결측 여부와 목표변수 상관관계")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

count_df = ttrain_.groupby(['missing_count', '허위매물여부']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 15))
count_df.plot(kind='bar', stacked=False, width=0.7, color=['black', 'red'])

# 그래프 스타일 설정
plt.xlabel('결측치수')
plt.ylabel('샘플 개수')
plt.title('결측치 개수에 따른 허위매물 여부 카운트')
plt.legend(['정상 매물(0)', '허위 매물(1)'])
plt.xticks(rotation=0)  # X축 라벨 가독성 조정
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# missing_count별 허위매물 비율 계산
count_df = ttrain_.groupby('missing_count')['허위매물여부'].mean().reset_index()

# 허위매물 비율 그래프 그리기
plt.figure(figsize=(8, 5))
sns.lineplot(data=count_df, x='missing_count', y='허위매물여부', marker='o', linestyle='-', color='red')

plt.xlabel("결측 개수 (missing_count)")
plt.ylabel("허위매물 비율")
plt.title("결측 개수에 따른 허위매물 비율 변화")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

결측치 개수 5에서는 허위매물이 없었지만, 결측치 개수에 따라 점차 허위매물 비율이 늘어나는 것이 보인다. 따라서 결측치 개수 컬럼을 추가한다.
